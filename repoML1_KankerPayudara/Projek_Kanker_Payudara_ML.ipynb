```python
import pandas as pd

# 1. Menentukan nama-nama kolom secara manual berdasarkan deskripsi wdbc.names
# Kolom 1: ID Pasien, Kolom 2: Diagnosis (M/B), diikuti oleh 30 fitur klinis sel
column_names = [
    'id', 'diagnosis',
    'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
    'compactness_mean', 'concavity_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean',
    'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
    'compactness_se', 'concavity_se', 'concave_points_se', 'symmetry_se', 'fractal_dimension_se',
    'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst',
    'compactness_worst', 'concavity_worst', 'concave_points_worst', 'symmetry_worst', 'fractal_dimension_worst'
]

# 2. Membaca file wdbc.data dan memasukkan nama kolom yang sudah dibuat
df = pd.read_csv('wdbc.data', header=None, names=column_names)

# 3. Menampilkan 5 baris pertama untuk memastikan data rapi dan terbaca
print("--- 5 BARIS PERTAMA DATASET ---")
print(df.head())

# 4. Menampilkan dimensi data (Jumlah baris dan kolom)
print("\n--- DIMENSI DATASET ---")
print(f"Jumlah Baris (Pasien): {df.shape[0]}")
print(f"Jumlah Kolom (Fitur): {df.shape[1]}")
```

The `FileNotFoundError` occurs because the `wdbc.data` file is not present in the current Colab environment. I will download it from a public UCI Machine Learning Repository URL.

```python
import requests

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'
filename = 'wdbc.data'

response = requests.get(url)
response.raise_for_status() # Raise an exception for HTTP errors

with open(filename, 'wb') as f:
    f.write(response.content)

print(f"'{filename}' downloaded successfully.")
```

The `wdbc.data` file has now been downloaded to the Colab environment. The code below will now execute without the `FileNotFoundError`.

```python
# 1. Menghapus kolom 'id' karena tidak memiliki nilai prediktif untuk model ML
if 'id' in df.columns:
    df = df.drop(columns=['id'])
    print("✓ Kolom 'id' berhasil dihapus.")

# 2. Memeriksa Missing Values (Data Kosong)
missing_values = df.isnull().sum().sum()
print(f"✓ Jumlah Missing Values di dataset: {missing_values}")

# 3. Memeriksa Distribusi Kelas Target (Diagnosis)
print("\n--- Distribusi Kelas Target ---")
print(df['diagnosis'].value_counts())

# 4. Memeriksa Persentase Distribusi Kelas Target
print("\n--- Persentase Distribusi Kelas Target ---")
print(df['diagnosis'].value_counts(normalize=True) * 100)
```

```python
from sklearn.model_selection import train_test_split

# 1. Encoding: Mengubah 'M' menjadi 1 dan 'B' menjadi 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
print("✓ Label target berhasil diubah menjadi angka (M=1, B=0).")

# 2. Memisahkan Fitur (X) dan Target (y)
X = df.drop(columns=['diagnosis']) # Semua kolom kecuali diagnosis adalah fitur
y = df['diagnosis']                # Kolom diagnosis adalah target

# 3. Data Splitting: Membagi menjadi 80% Train dan 20% Test
# stratify=y digunakan agar proporsi kelas M dan B di data train & test tetap seimbang (80:20)
# random_state=42 digunakan agar hasil pembagian data selalu konsisten saat dijalankan ulang
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("\n--- HASIL PEMBAGIAN DATA ---")
print(f"Jumlah Data Latih (X_train): {X_train.shape[0]} sampel")
print(f"Jumlah Data Uji (X_test): {X_test.shape[0]} sampel")
print(f"Jumlah Fitur per Sampel: {X_train.shape[1]} fitur")
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Inisialisasi Objek StandardScaler
scaler = StandardScaler()

# 2. Fit dan Transform pada Data Latih (X_train)
# Model akan mempelajari rata-rata dan standar deviasi hanya dari data latih
X_train_scaled = scaler.fit_transform(X_train)

# 3. Transform pada Data Uji (X_test) menggunakan parameter yang sudah dipelajari
# PERINGATAN: Cukup gunakan .transform(), JANGAN gunakan .fit_transform() di data uji agar tidak terjadi data leakage!
X_test_scaled = scaler.transform(X_test)

print("✓ Proses Fitur Scaling berhasil diselesaikan!")
print("\nContoh Perbandingan Data Sebelum dan Sesudah Scaling (Baris Pertama):")
print("-" * 65)
print(f"Sebelum Scaling (X_train) :\n{X_train.iloc[0][:3].values} ... (dan seterusnya)")
print(f"\nSesudah Scaling (X_train_scaled) :\n{X_train_scaled[0][:3]} ... (dan seterusnya)")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# 1. Inisialisasi Model DWKNN
# n_neighbors=5 (melihat 5 tetangga terdekat)
# weights='distance' (mengaktifkan rumus pembobotan berdasarkan jarak)
model_dwknn = KNeighborsClassifier(n_neighbors=5, weights='distance')

# 2. Melakukan 5-Fold Cross Validation menggunakan Data Latih
# Metrik yang digunakan adalah 'accuracy' untuk melihat performa umum dulu
scores = cross_val_score(model_dwknn, X_train_scaled, y_train, cv=5, scoring='accuracy')

# 3. Melatih model secara penuh pada data latih (untuk digunakan nanti)
model_dwknn.fit(X_train_scaled, y_train)

print("✓ Model DWKNN berhasil diinisialisasi dan dilatih!")
print("\n--- HASIL 5-FOLD CROSS VALIDATION ---")
print(f"Akurasi tiap Lipatan (Fold) : {scores}")
print(f"Rata-rata Akurasi (Mean Acc) : {np.mean(scores) * 100:.2f}%")
print(f"Standar Deviasi Akurasi      : {np.std(scores) * 100:.2f}%")
```

```python
from sklearn.neural_network import MLPClassifier

# 1. Inisialisasi Model ANN (Multi-Layer Perceptron)
# hidden_layer_sizes=(16, 8) -> Membuat 2 hidden layer (layer pertama 16 neuron, layer kedua 8 neuron)
# max_iter=500 -> Memberikan kesempatan model untuk belajar hingga 500 iterasi (epochs)
# random_state=42 -> Agar hasil latihan konsisten saat dijalankan ulang
model_ann = MLPClassifier(hidden_layer_sizes=(16, 8), max_iter=500, random_state=42)

# 2. Melakukan 5-Fold Cross Validation menggunakan Data Latih yang sama
scores_ann = cross_val_score(model_ann, X_train_scaled, y_train, cv=5, scoring='accuracy')

# 3. Melatih model secara penuh pada data latih
model_ann.fit(X_train_scaled, y_train)

print("✓ Model ANN berhasil diinisialisasi dan dilatih!")
print("\n--- HASIL 5-FOLD CROSS VALIDATION (ANN) ---")
print(f"Akurasi tiap Lipatan (Fold) : {scores_ann}")
print(f"Rata-rata Akurasi (Mean Acc) : {np.mean(scores_ann) * 100:.2f}%")
print(f"Standar Deviasi Akurasi      : {np.std(scores_ann) * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, f1_score

# 1. Prediksi menggunakan kedua model pada Data Uji Tersembunyi
y_pred_dwknn = model_dwknn.predict(X_test_scaled)
y_pred_ann = model_ann.predict(X_test_scaled)

# 2. Menampilkan Laporan Evaluasi untuk Model DWKNN
print("========================================================")
print("       LAPORAN EVALUASI FINAL: MODEL DWKNN              ")
print("========================================================")
print(f"Akurasi Final  : {accuracy_score(y_test, y_pred_dwknn) * 100:.2f}%")
print(f"Recall Ganas   : {recall_score(y_test, y_pred_dwknn) * 100:.2f}%")
print(f"F1-Score Ganas : {f1_score(y_test, y_pred_dwknn) * 100:.2f}%")
print("\nDetail Metrik Lengkap:")
print(classification_report(y_test, y_pred_dwknn, target_names=['Benign (0)', 'Malignant (1)']))

# 3. Menampilkan Laporan Evaluasi untuk Model ANN
print("\n========================================================")
print("       LAPORAN EVALUASI FINAL: MODEL ANN                ")
print("========================================================")
print(f"Akurasi Final  : {accuracy_score(y_test, y_pred_ann) * 100:.2f}%")
print(f"Recall Ganas   : {recall_score(y_test, y_pred_ann) * 100:.2f}%")
print(f"F1-Score Ganas : {f1_score(y_test, y_pred_ann) * 100:.2f}%")
print("\nDetail Metrik Lengkap:")
print(classification_report(y_test, y_pred_ann, target_names=['Benign (0)', 'Malignant (1)']))
```

```python
import joblib

# 1. Menyimpan model DWKNN yang sudah dilatih
joblib.dump(model_dwknn, 'model_dwknn.pkl')
print("✓ Model DWKNN berhasil disimpan menjadi 'model_dwknn.pkl'")

# 2. Menyimpan model ANN yang sudah dilatih
joblib.dump(model_ann, 'model_ann.pkl')
print("✓ Model ANN berhasil disimpan menjadi 'model_ann.pkl'")

# 3. Menyimpan objek Scaler (sangat penting untuk menyamakan data baru nanti)
joblib.dump(scaler, 'scaler.pkl')
print("✓ Scaler berhasil disimpan menjadi 'scaler.pkl'")
```
